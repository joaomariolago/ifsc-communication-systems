# Circuito Modulador

## Esquemático

![](./prints/modulator_circuit.png)

O circuito modulador é composto pela soma de sinais através de resistores, seguida pela atuação de um diodo que funciona como uma chave síncrona. Este processo gera um sinal modulado contendo harmônicos, que posteriormente são atenuados por um filtro.

## Simulação


### Sinais de entrada 

![](./prints/modulator_inputs.png)

Conforme observado nos sinais de entrada, a frequência predominante do sinal é de 1 kHz, com amplitude variando para cada valor de índice de modulação $ M $ [0,1; 0,5; 0,9]. O sinal portadora (*carrier*) mantém sua amplitude constante em 100 kHz.

**Nota:** há uma pequena influência do sinal de entrada sobre a portadora, devido ao fato de que o simulador utilizado (LTspice) considera não idealidades dos componentes, como resistência interna. Isso gera uma interação mútua entre os sinais.

### Análise dos sinais do circuito

![](./prints/modulator_outputs.png)

- No sinal *carrier + signal* (ponto 3), os resistores realizam a divisão e a soma dos sinais de entrada.
- No sinal *half_wave* (ponto 4), o diodo conduz apenas a parte positiva da onda, gerando a modulação acompanhada de harmônicos.
- No sinal *output* (ponto 5), o filtro atua atenuando as harmônicas localizadas longe da frequência da portadora.

#### Efeito do índice de modulação

Com o aumento do índice de modulação $M$, o sinal modulado apresenta maior distorção, e o filtro passa a ter dificuldade para eliminar frequências indesejadas, permitindo a passagem de componentes fora da banda da portadora.

##### M = 0,1

![](./prints/modulator_outputs_m_100m.png)

Neste caso, os harmônicos permanecem controlados. Observa-se a frequência da portadora e as duas bandas laterais, distanciadas pela frequência do sinal de entrada (*signal*), características da modulação. Os harmônicos de ordens superiores possuem potência desprezível.

##### M = 0,5

![](./prints/modulator_outputs_m_500m.png)

Com o aumento de $ M $, começam a surgir harmônicos múltiplos, visíveis até aproximadamente a sexta ordem da frequência do sinal, resultando em uma leve distorção, ainda pouco perceptível no domínio do tempo.

##### M = 0,9

![](./prints/modulator_outputs_m_900m.png)

Com $ M $ próximo de 1, o circuito apresenta distorções significativas, evidentes tanto na análise temporal quanto na análise de frequência, onde a energia dos harmônicos se aproxima da energia da portadora.

### Comparação com o sinal esperado

Considerando que nas análises anteriores o circuito demonstrou melhor desempenho para $M$ próximo a 0,1, as próximas comparações serão feitas apenas com $ M = 0,1 $.

![](./prints/modulator_expected_signals.png)

![](./prints/modulator_all_signals_100m.png)

- **V1** e **V2**, correspondentes ao sinal da portadora e ao sinal de entrada, respectivamente, mantêm o mesmo padrão, com a principal diferença sendo a amplitude. Devido ao uso de modelos realistas no simulador, a baixa amplitude causa perdas de sinal pela queda de tensão no diodo.
- **V3** apresenta a soma dos dois sinais de maneira satisfatória.
- **V4** representa corretamente a condução apenas da metade da onda modulada (meia onda positiva).
- **V5** corresponde ao sinal modulado final, após o processo de filtragem.

De maneira geral, os sinais se comportam conforme o esperado, com a ressalva de que foi necessário trabalhar com amplitudes maiores para compensar as limitações introduzidas pelas características reais do diodo utilizado no circuito.


# Circuito Demodulador

## Cálculos

Para o dimensionamento dos componentes, será utilizada a seguinte fórmula:

$$
\frac{1}{RC} \geq m \cdot 2\pi f_{\text{máx}}
$$

O maior valor de $ m $ utilizado no modulador é 0,9, representando o pior caso. Portanto, esse valor será considerado nos cálculos a seguir, garantindo que o demodulador funcione adequadamente para todas as simulações realizadas.

Isolando $ R $:

$$
R \leq \frac{1}{C \cdot m \cdot 2\pi f_{\text{máx}}}
$$

Assumindo a utilização de um capacitor de 100 nF:

$$
R \leq \frac{1}{100\ \text{nF} \cdot 0{,}9 \cdot 2\pi \cdot 1 \times 10^{3}}
$$

Calculando:

$$
R \leq 1{,}7\ \text{k}\Omega
$$

## Simulação

![](./prints/demodulator_simulation.png)

Como é possível observar na simulação, o circuito demodulador consegue recuperar o sinal de entrada original. No entanto, ainda permanecem algumas harmônicas de alta frequência, embora com amplitudes significativamente reduzidas em comparação ao sinal modulado.
